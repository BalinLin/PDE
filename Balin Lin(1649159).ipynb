# Exercise 1

```python
def set_u0(x):
    if 0.4 <= x < 0.6:
        return 4
    else:
        return 1
```

```python
N = 200
T = 0.01

a = 1
b = 0
c = 1

h = (a - b) / N

dt = h*h / (2 * c) * 0.9

space = np.linspace(a, b, N)

u0 = np.array(list(map(set_u0, space)))
u1 = np.zeros(u0.shape)

boundary = 'Neumann'
```

```python
t = 0
while t < T:
    for i in range(1, N - 1):
        u1[i] = u0[i] + c * (dt / (h*h)) * (u0[i-1] - 2*u0[i] + u0[i+1])
        if boundary == 'Dirichlet':
            u1[0] = u0[0]
            u1[N-1] = u0[N-1]
        elif boundary == 'Neumann':
            u1[0] = u0[1]
            u1[N - 1] = u0[N - 2]

    u0 = u1
    t += dt
    
plt.title(f" Homogeneous {boundary}. N={N}, T={T}")
plt.plot(u0)
plt.show()
```

# Exercise 2

```python
lam = 1
n = 10
B = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == 0 and j == n-1:
            B[i][j] = (-0.5)*lam
        elif i == n-1 and j == 0:
            B[i][j] = (-0.5)*lam
        elif i == j:
            B[i][j] = 1 + lam
        elif i == j-1:
            B[i][j] = (-0.5)*lam
        elif i-1 == j:
            B[i][j] = (-0.5)*lam
            
print("B:\n", B)

D = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == 0 and j == n-1:
            D[i][j] = (0.5)*lam
        elif i == n-1 and j == 0:
            D[i][j] = (0.5)*lam
        elif i == j:
            D[i][j] = 1 - lam
        elif i == j-1:
            D[i][j] = (0.5)*lam
        elif i-1 == j:
            D[i][j] = (0.5)*lam
            
print("D:\n", D)
```

```python
u1 = np.eye(n)
u = np.eye(n)

x = B @ u1
y = D @ u

print("x:\n", x)
print("y:\n", y)

ans = np.linalg.solve(x, y)
print("ans:\n", ans)
```

# Exercise 3

```python
n = 200
delta_x = 2 / n
x = [0] * n
for i in range(n):
    x[i] = -1 + i * delta_x

print("x")
print(x)
```

```python
phi_u = [0] * n
for i in range(n):
    if abs(x[i]) <= 0.5:
        phi_u[i] = 1

print("phi_u:")
print(phi_u)
```

```python
plt.title("Line graph")
plt.plot(phi_u, color="red")

plt.show()
```

```python
CFL = 0.8
c = 1
delta_t = delta_x * CFL
time = 100
method = 0.1

phi_u = [0] * n
for i in range(n):
    if abs(x[i]) <= 0.5:
        phi_u[i] = 1
        
for t in range(time):
    new_u = list(phi_u)
    for i in range(0, n):
        new_u[i] = phi_u[i-1] - c * (delta_t / delta_x) * (method)
    phi_u = new_u
    plt.title("Line graph")
    plt.plot(new_u, color="red")
    plt.show()
```
